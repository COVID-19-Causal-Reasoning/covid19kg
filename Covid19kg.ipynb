# BEL language specification

```python
import yaml
!wget -c https://raw.githubusercontent.com/belbio/bel_specifications/master/specifications/bel_v2_1_0.yaml
```

```python
with open("bel_v2_1_0.yaml", 'r') as stream:
    try:
        bel_spec = yaml.safe_load(stream)
        print(bel_spec)
    except yaml.YAMLError as exc:
        print(exc)
```

```python
from collections import defaultdict
function_categories = defaultdict(list)
info = bel_spec['functions']['info']
for func in info:
    for category in info[func]['categories']:
        function_categories[category].append(func)
```
